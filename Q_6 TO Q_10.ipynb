# QUESTION-6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both of the lines and show the legend.

Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt
us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = us_marriage_divorce_data['Year'].values
#print(years)
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values
plt.plot(years,marriages_per_capita, years,divorces_per_capita)
plt.title('line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014')
plt.xlabel('Years')
plt.ylabel('marriages_per_capita and divorces_per_capita')
plt.legend(["blue", "green"], loc ="lower right")
plt.show()

```

# Question-7

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.

Don't forget to label your axes!

```python
import pandas as pd
import numpy as np

us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us_marriage_divorce_data = us_marriage_divorce_data[
us_marriage_divorce_data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = us_marriage_divorce_data['Year'].values
print(years)
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

X = np.arange(len(years))
# Passing the parameters to the bar function, this is the main function which creates the bar plot
# Using X now to align the bars side by side
plt.bar(X,marriages_per_capita  ,color = 'black', width = 0.25)
plt.bar(X + 0.25, divorces_per_capita , color = 'g', width = 0.25)
#plt.bar(X + 0.25, years, color = 'b', width = 0.25)
plt.legend(['marriages_per_capita', 'divorces_per_capita'])

plt.xticks([i + 0.25 for i in range(3)], [1900, 1950, 2000])

plt.title("Bar plot representing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000")
# Namimg the x and y axis
plt.xlabel('Years')
plt.ylabel('Count of marraiges/divorces ')

plt.show()

```

# Question-8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.

Don't forget to label your axes!

For reference, here's the raw data:

```python
import pandas as pd
import matplotlib.pyplot as plt
hollywood_actor_kills = pd.read_csv('actor_kill_counts.csv')
hollywood_actor_kills_descending = hollywood_actor_kills.sort_values('Count',ascending=False)

actors = hollywood_actor_kills_descending['Actor'].values
kills = hollywood_actor_kills_descending['Count'].values
print(kills)

plt.barh(actors,kills)
plt.title('Horizontal Bar Plot of Actors and their kill count')
plt.ylabel('Actors')
plt.xlabel('Kill Count')
plt.show()
```

# Question-9

Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
roman_emperors = pd.read_csv('roman-emperor-reigns.csv')
assassinated_emperors = roman_emperors[
    roman_emperors['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]


number_assassinated = len(assassinated_emperors)
other_deaths = len(roman_emperors) - number_assassinated


emperor = assassinated_emperors["Emperor"]
cause_of_death = assassinated_emperors["Cause_of_Death"]
plt.pie(range(len(cause_of_death)), labels=emperor,autopct='%1.1f%%', startangle=90, radius=0.04 * 70,rotatelabels = 270)
fig = plt.figure(figsize=[2, 2])
```
